```python
#Running TPOT - running version 0.3.0
import pandas as pd
from sklearn.cross_validation import train_test_split  
from tpot import TPOT  

#Import dataset
data = pd.read_csv('/home/ubuntu/notebooks/iris.data', sep=',', header = None)

#Split features and labels
X = pd.DataFrame(data, columns=[0,1,2,3])
y = pd.DataFrame(data, columns=[4])
#y = labels.replace({"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2})

#Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)  

#Fit data using TPOT
my_tpot = TPOT(generations=10)  
my_tpot.fit(X_train, y_train)  

#Average crossvalidated accuracy
print(my_tpot.score(X_test, y_test))
```
