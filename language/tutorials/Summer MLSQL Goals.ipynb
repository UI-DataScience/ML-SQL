### Keywords Implemented

- LOAD - load pre-trained model from memory
- READ - read in data file
- SPLIT - split data file into training, testing, and validation sets
- REPLACE - replace missing values in a dataset
- CLASSIFY - perform classification
- REGRESSION - perform regression
- CLUSTER - perform clustering
- APPLY - label data based on the model
- SAVE - Used to save a model to a file
